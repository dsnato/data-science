```python
# Traduzir nomes das colunas para portuguÃªs
df.columns = ['ano_trabalho', 'nivel_experiencia', 'tipo_emprego', 'cargo',
              'salario', 'moeda_salario', 'salario_em_usd', 'residencia_empregado',
              'proporcao_remoto', 'localizacao_empresa', 'tamanho_empresa']

# Exibir os nomes das colunas traduzidas para verificar
display(df.columns)
```

```python
# Mapear abreviaÃ§Ãµes para nomes completos em portuguÃªs
mapeamento_nivel_experiencia = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

# Aplicar o mapeamento Ã  coluna 'nivel_experiencia'
df['nivel_experiencia'] = df['nivel_experiencia'].map(mapeamento_nivel_experiencia)

# Exibir os valores Ãºnicos para verificar
display(df['nivel_experiencia'].unique())
```

```python
# Mapear abreviaÃ§Ãµes para nomes completos em portuguÃªs
mapeamento_tipo_emprego = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

# Aplicar o mapeamento Ã  coluna 'tipo_emprego'
df['tipo_emprego'] = df['tipo_emprego'].map(mapeamento_tipo_emprego)

# Exibir os valores Ãºnicos para verificar
display(df['tipo_emprego'].unique())
```

```python
# Mapear categorias numÃ©ricas para descriÃ§Ãµes em portuguÃªs
mapeamento_proporcao_remoto = {
    0: 'Presencial',
    50: 'HÃ­brido',
    100: 'Remoto'
}

# Aplicar o mapeamento Ã  coluna 'proporcao_remoto'
df['proporcao_remoto'] = df['proporcao_remoto'].map(mapeamento_proporcao_remoto)

# Exibir os valores Ãºnicos para verificar
display(df['proporcao_remoto'].unique())
```
